## III.II.II. Code Explanation: Model-1
Using Keras, I developed a sequential neural network model in this code. The model is composed of an output layer, 5 hidden layers, each comprising 64 neurons, and an input layer with 5 neurons (matching the structure of the data).

After establishing the model, I built it using the Adam optimizer, with accuracy serving as the training metric and binary_crossentropy as the loss function.

I then used the training data to train the model. I also tracked the time spent on this procedure as an extra credit in order to gauge how long it took to develop and train the model.

In conclusion, the code generates a sequential neural network model, describes its architecture, compiles it with the selected optimizer, loss function, and metric, trains it with the data, and then the training data and keeps track of how long this process took.


```python
# Start measuring time
start_time = timeit.default_timer()

# Define the sequential model with 5 hidden layers
model_1 = keras.Sequential([
        layers.Dense(units=5, input_dim=5, activation="tanh", name="input"),
        layers.Dense(units=64, activation="tanh", name="hidden_1"),
        layers.Dense(units=64, activation="tanh", name="hidden_2"),
        layers.Dense(units=64, activation="tanh", name="hidden_3"),
        layers.Dense(units=64, activation="tanh", name="hidden_4"),
        layers.Dense(units=64, activation="tanh", name="hidden_5"),
        layers.Dense(units=1, activation="tanh", name="output"),
    ])

# Define the Adam optimizer
opt_adam = optimizers.Adam(learning_rate=0.001)

# Compile the model
model_1.compile(optimizer=opt_adam, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model on the training data
res_1 = model_1.fit(x=X_train, y=T_train, epochs=20, batch_size=1000)

# Record the time taken to build and train the model
elapsed = timeit.default_timer() - start_time
print("Time taken by Model-1: {} sec.".format(elapsed))

# Print a summary of the model's architecture
print(model_1.summary())
```

## III.II.III. Code Explanation: Model-2
The Keras Sequential model is used in this code to generate a neural network. The network consists of an output layer, 5 hidden layers, and an input layer of 5 neurons. There are 64 neurons in the first hidden layer, and 64 more neurons are added in each additional layer after that. The "tanh" function is the activation function utilized in all layers.

The binary cross-entropy loss function, accuracy, and the Adam optimizer are used in the model's construction. The fit approach is then used to train it using the training data. The code also keeps track of how long it took to create and train the model.


```python
start_time = timeit.default_timer()

# Define Sequential model with 5 hidden layers
model_2 = keras.Sequential(
    [
        layers.Dense(units=5, input_dim=5, activation="tanh", name="input"),
        layers.Dense(units=64, activation="tanh", name="hidden_1"),
        layers.Dense(units=128, activation="tanh", name="hidden_2"),
        layers.Dense(units=256, activation="tanh", name="hidden_3"),
        layers.Dense(units=512, activation="tanh", name="hidden_4"),
        layers.Dense(units=1024, activation="tanh", name="hidden_5"),
        layers.Dense(units=1, activation="tanh", name="output"),
    ]
)

# compile the keras model
model_2.compile(opt_adam, loss='binary_crossentropy', metrics=['accuracy'])

# fit the keras model on the dataset
res_2 = model_2.fit(X_train, T_train, epochs=20, batch_size=1000)

elapsed = timeit.default_timer() - start_time
print("Time taken by Model-2: {} sec.".format(elapsed))

print(model_2.summary())
```

## III.II.IV. Code Explanation: Model-3
This model employs the Sequential model from Keras and has a 5-neuron input layer that reflects the data's form. There are nine hidden levels altogether, each with a different amount of neurons than the models before it. The first hidden layer contains 64 neurons, while the following four layers each contain twice as many neurons as the one before it. The number of neurons is reduced by a factor of 2x for each of the remaining hidden layers. For all layers, the activation function is tanh. With accuracy serving as the evaluation measure and binary_crossentropy serving as the loss function, the model is built using the Adam optimizer. After utilizing the training data to train the model,and for extra credit, the model's construction and training times are measured.



```python
start_time = timeit.default_timer()

# Define Sequential model with 9 hidden layers
model_3 = keras.Sequential(
    [
        layers.Dense(units=5, input_dim=5, activation="tanh", name="input"),
        layers.Dense(units=64, activation="tanh", name="hidden_1"),
        layers.Dense(units=128, activation="tanh", name="hidden_2"),
        layers.Dense(units=256, activation="tanh", name="hidden_3"),
        layers.Dense(units=512, activation="tanh", name="hidden_4"),
        layers.Dense(units=1024, activation="tanh", name="hidden_5"),
        layers.Dense(units=512, activation="tanh", name="hidden_6"),
        layers.Dense(units=256, activation="tanh", name="hidden_7"),
        layers.Dense(units=128, activation="tanh", name="hidden_8"),
        layers.Dense(units=64, activation="tanh", name="hidden_9"),
        layers.Dense(units=1, activation="tanh", name="output"),
    ]
)

# compile the keras model
model_3.compile(opt_adam, loss='binary_crossentropy', metrics=['accuracy'])

# fit the keras model on the dataset
res_3 = model_3.fit(X_train, T_train, epochs=20, batch_size=1000)

elapsed = timeit.default_timer() - start_time
print("Time taken by Model-3: {} sec.".format(elapsed))

print(model_3.summary())
```

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate a toy binary classification dataset
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=123)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Define the hyperparameters
input_size = X_train.shape[1]
hidden_size1 = 32
hidden_size2 = 16
hidden_size3 = 8
hidden_size4 = 4
output_size = 2
learning_rate = 0.1
num_iterations = 1000

# Initialize the parameters
params = {
    'W1': np.random.randn(input_size, hidden_size1) * 0.01,
    'b1': np.zeros(hidden_size1),
    'W2': np.random.randn(hidden_size1, hidden_size2) * 0.01,
    'b2': np.zeros(hidden_size2),
    'W3': np.random.randn(hidden_size2, hidden_size3) * 0.01,
    'b3': np.zeros(hidden_size3),
    'W4': np.random.randn(hidden_size3, hidden_size4) * 0.01,
    'b4': np.zeros(hidden_size4),
    'W5': np.random.randn(hidden_size4, output_size) * 0.01,
    'b5': np.zeros(output_size)
}
def softmax(z):
    z -= np.max(z)
    sm = (np.exp(z).T / np.sum(np.exp(z), axis=1)).T
    return sm

# Train the model
for i in range(num_iterations):
    # Forward propagation
    Z1 = np.dot(X_train, params['W1']) + params['b1']
    A1 = np.tanh(Z1)
    
    Z2 = np.dot(A1, params['W2']) + params['b2']
    A2 = np.tanh(Z2)
    
    Z3 = np.dot(A2, params['W3']) + params['b3']
    A3 = np.tanh(Z3)
    
    Z4 = np.dot(A3, params['W4']) + params['b4']
    A4 = np.tanh(Z4)
    
    Z5 = np.dot(A4, params['W5']) + params['b5']
    A5 = softmax(Z5)
    
    # Compute the cross-entropy cost
    m = y_train.shape[0]
    Y = np.zeros((m, output_size))
    Y[np.arange(m), y_train] = 1
    
    logprobs = np.multiply(-np.log(A5), Y)
    cost = np.sum(logprobs) / m
    
    # Backward propagation
    dZ5 = A5 - Y
    dW5 = 1/m * np.dot(A4.T, dZ5)
    db5 = 1/m * np.sum(dZ5, axis=0)

    dA4 = np.dot(dZ5, params['W5'].T)
    dZ4 = np.multiply(dA4, 1 - np.power(A4, 2))
    dW4 = 1/m * np.dot(A3.T, dZ4)
    db4 = 1/m * np.sum(dZ4, axis=0)

    dA3 = np.dot(dZ4, params['W4'].T)
    dZ3 = np.multiply(dA3, 1 - np.power(A3, 2))
    dW3 = 1/m * np.dot(A2.T, dZ3)
    db3 = 1/m * np.sum(dZ3, axis=0)
```

```python
dA2 = np.dot(dZ3, params['W3'].T)
dZ2 = np.multiply(dA2, 1 - np.power(A2, 2))
dW2 = 1/m * np.dot(A1.T, dZ2)
db2 = 1/m * np.sum(dZ2, axis=0)

dA1 = np.dot(dZ2, params['W2'].T)
dZ1 = np.multiply(dA1, 1 - np.power(A1, 2))
dW1 = 1/m * np.dot(X_train.T, dZ1)
db1 = 1/m * np.sum(dZ1, axis=0)

# Update the parameters
params['W5'] -= learning_rate * dW5
params['b5'] -= learning_rate * db5
params['W4'] -= learning_rate * dW4
params['b4'] -= learning_rate * db4
params['W3'] -= learning_rate * dW3
params['b3'] -= learning_rate * db3
params['W2'] -= learning_rate * dW2
params['b2'] -= learning_rate * db2
params['W1'] -= learning_rate * dW1
params['b1'] -= learning_rate * db1
```

```python
#Make predictions on the test set
Z1 = np.dot(X_test, params['W1']) + params['b1']
A1 = np.tanh(Z1)

Z2 = np.dot(A1, params['W2']) + params['b2']
A2 = np.tanh(Z2)

Z3 = np.dot(A2, params['W3']) + params['b3']
A3 = np.tanh(Z3)

Z4 = np.dot(A3, params['W4']) + params['b4']
A4 = np.tanh(Z4)

Z5 = np.dot(A4, params['W5']) + params['b5']
A5 = softmax(Z5)

y_pred = np.argmax(A5, axis=1)

#Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
